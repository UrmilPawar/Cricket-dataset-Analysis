### Q 18. Percentage runs and wickets for teams 
#### England ( till 17)
#### India ( after 17)

```python
df = pd.read_csv('combined.csv')
players=df['Player'].iloc[0:18]
runs=df['runs_scored'].iloc[0:18]
wickets=df['wickets'].iloc[0:18]
sum1=runs.sum()
sum2=wickets.sum()
df['per_runs'] = runs.apply(lambda runs:(runs/sum1)*100)
df['per_wckt'] = wickets.apply(lambda wickets:(wickets/sum2)*100)
print("The percentage runs and wickets for England is" )
df.iloc[0:18]
```

```python
df = pd.read_csv('combined.csv')
players=df['Player'].iloc[18:31]
runs=df['runs_scored'].iloc[18:31]
wickets=df['wickets'].iloc[18:31]
sum1=runs.sum()
sum2=wickets.sum()
df['per_runs'] = runs.apply(lambda runs:(runs/sum1)*100)
df['per_wckt'] = wickets.apply(lambda wickets:(wickets/sum2)*100)
print("The percentage runs and wickets for India is" )
df.iloc[18:31]

```

### Q19 . Plot a pie chart for displaying the England run percentage

```python
df = pd.read_csv('combined.csv')
plt.figure(figsize=(15,25))
players=df['Player'].iloc[0:18]
runs=df['runs_scored'].iloc[0:18]
plt.pie(runs,labels=players,autopct='%.2f%%')
plt.show()
```
